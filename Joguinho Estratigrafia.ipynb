```python
import numpy
import random
```

```python
# Listas das idades absolutas e respectivas idades relativas:

Maast = list(numpy.arange(66,72.1,0.1)) 
Camp = list(numpy.arange(72.1,83.6,0.1)) 
Santo = list(numpy.arange(83.6,86.3,0.1)) 
Coniac = list(numpy.arange(86.3,89.8,0.1)) 
Turo = list(numpy.arange(89.8,93.9,0.1)) 
Ceno = list(numpy.arange(93.9,100.5,0.1)) 

# Arredondando os valores float das listas:

Maast_cor= list()
for i in Maast:
    j=round(i,1)
    Maast_cor.append(j)
Camp_cor= list()
for i in Camp:
    j=round(i,1)
    Camp_cor.append(j)
Santo_cor= list()
for i in Santo:
    j=round(i,1)
    Santo_cor.append(j)    
Coniac_cor= list()
for i in Coniac:
    j=round(i,1)
    Coniac_cor.append(j)    
Turo_cor= list()
for i in Turo:
    j=round(i,1)
    Turo_cor.append(j)    
Ceno_cor= list()
for i in Ceno:
    j=round(i,1)
    Ceno_cor.append(j)  
        
# Entrada do jogo:

jogador = input(str("Olá, qual seu nome: "))
pontuacao = 0
print('')
print('{}, tente acertar os períodos do Cretáceo Superior correspondentes às 5 idades absolutas '.format(jogador))
print('que o programa vai gerar.')
print('')
print('Para responder, digite apenas as siglas da legenda abaixo:')
print('')

# Legenda para o jogador:
print('LEGENDA:')        
print('   Maastrichtiano = M')
print('   Campaniano = Ca')
print('   Santoniano = S')
print('   Coniaciano = Co')
print('   Turoniano = T')
print('   Cenomaniano = Ce')
print('')

# Gerando uma idade qualquer e atribuindo a respectiva idade relativa:

for x in range(1,6):
    age = round(random.uniform(66,100.5),1)

    for i in Maast_cor:
        if i == age:
            gabarito = "M"         
    for i in Camp_cor:
        if i == age:
            gabarito = "Ca"        
    for i in Santo_cor:
        if i == age:
            gabarito = "S"   
    for i in Coniac_cor:
        if i == age:
            gabarito = "Co"
    for i in Turo_cor:
        if i == age:
            gabarito = "T"        
    for i in Ceno_cor:
        if i == age:
            gabarito = "Ce"
            
    # Comparando resposta do jogador com a idade gerada:
    
    idadex = input('{}: Qual período da cronoestratigrafia pertence a idade de {}.Ma? '.format(x,age))
    if idadex == gabarito:
        print("Certa resposta!",gabarito)
        pontuacao+=1
    else:
        print("Erroooou!!",gabarito)
print('')
print("Resultado: O jogador {} conseguiu atingir {} pontos em 5 tentativas.".format(jogador,pontuacao))
```
